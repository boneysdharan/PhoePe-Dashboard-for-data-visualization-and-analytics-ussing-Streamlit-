```python
#This is for aggregated transtaction state wise data
import pandas as pd
import json
import os

#This is to direct the path to get the data as states

path=r"D:/BONEYS/WEB/PYTHON/Project/PhonePe/data/aggregated/transaction/country/india/state"
Agg_state_list=os.listdir(path)
Agg_state_list

#This is to extract the data's to create a dataframe
clm={'State':[], 'Year':[],'Quater':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in Agg_state_list:
    p_i=path+"/"+i
    Agg_yr=os.listdir(p_i)
    #delhi
    for j in Agg_yr:
        p_j=p_i+"/"+j
        #2018
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+"/"+k 
            #1.json
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clm['Transaction_type'].append(Name)
              clm['Transaction_count'].append(count)
              clm['Transaction_amount'].append(amount)
              clm['State'].append(i)
              clm['Year'].append(j)
              clm['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Agg_Trans_state=pd.DataFrame(clm)
print(Agg_Trans_state)
```

```python
#creating table to store the above data
import mysql.connector
connection = mysql.connector.connect(
    host="localhost",
    user="xxx", #give your user name. Mostly its root
    password="xxx", #give your password
    database="xxx" #give any database name you want
)
cursor=connection.cursor()
```

```python
query1 = """
        create table Aggregated_Transaction_State(
        State varchar(100),
        Year int,
        Quater int,
        Transaction_Type varchar(100),
        Transaction_count float,
        transaction_amount float)
        """
cursor.execute(query1)
```

```python
#storing the above in mysql database
#all data must be tuple and must be a list of tuples
data=[]
for index in Agg_Trans_state.index:
    row = Agg_Trans_state.loc[index]
    row = row.values
    row = row[0],float(row[1]),int(row[2]),row[3],int(row[4]),float(row[5])
    data.append(row)
```

```python
query2 = "insert into Aggregated_Transaction_State values (%s,%s,%s,%s,%s,%s)"
cursor.executemany(query2,data)
connection.commit()
```

```python
#doing data cleaning for Agg_Trans_state
from sklearn.preprocessing import LabelEncoder
#first step check and remove null values
Agg_Trans_state.isna().sum()
Agg_Trans_state.info()
#Second do encoding and dummies to make all values int
enc_col = ['Transaction_type']
dummy_col = ['Year','State','Quater']
encoder = LabelEncoder()
Agg_Trans_state['encoded_transaction_type'] = encoder.fit_transform(Agg_Trans_state['Transaction_type'])
Agg_Trans_state = pd.get_dummies(Agg_Trans_state,columns=dummy_col,dtype=int)
#third removing outliners
Agg_Trans_state['transaction_count_std']=(Agg_Trans_state['Transaction_count']-Agg_Trans_state['Transaction_count'].mean())/Agg_Trans_state['Transaction_count'].std()
Agg_Trans_state['transaction_amount_std']=(Agg_Trans_state['Transaction_amount']-Agg_Trans_state['Transaction_amount'].mean())/Agg_Trans_state['Transaction_amount'].std()
condition1 = Agg_Trans_state['transaction_count_std']<=3
condition2 = Agg_Trans_state['transaction_count_std']>=-3
condition3 = Agg_Trans_state['transaction_amount_std']<=3
condition4 = Agg_Trans_state['transaction_amount_std']>=-3
Agg_Trans_state = Agg_Trans_state[(condition1) & (condition2) & (condition3) & (condition4)]
print(Agg_Trans_state)
```

```python
#This is for aggregated transaction for the year 2018 data
import pandas as pd
import json
import os

#This is to direct the path to get the data as states

path=r"D:/BONEYS/WEB/PYTHON/Project/PhonePe/data/aggregated/transaction/country/india/2018"
Agg_tran_2018_list=os.listdir(path)

#This is to extract the data's to create a dataframe
clm={'Quater':[],'Transacion_type':[], 'Transacion_count':[], 'Transacion_amount':[]}

for i in Agg_tran_2018_list:
    p_i=path+"/"+i
    Data=open(p_i,'r')
    D=json.load(Data)
    for z in D['data']['transactionData']:
        Name=z['name']
        count=z['paymentInstruments'][0]['count']
        amount=z['paymentInstruments'][0]['amount']
        clm['Transacion_type'].append(Name)
        clm['Transacion_count'].append(count)
        clm['Transacion_amount'].append(amount)
        clm['Quater'].append(int(i.strip('.json')))
#Succesfully created a dataframe
Agg_Trans_2018=pd.DataFrame(clm)
print(Agg_Trans_2018)
```
